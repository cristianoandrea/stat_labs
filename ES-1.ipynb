```python
import numpy as np
import scipy.linalg
import matplotlib.pyplot as plt
```

```python
n_all= np.arange(10, 100, 10)
print(n_all)

condition=[]
condition_inf=[]
error=[]

for n in (n_all):

    #Computes the right-hand side of the linear system b = Axtrue.
    A= np.random.rand(n,n)
    x_true = np.ones(n)
    b= A@x_true

    print("A:", A, '\n b:', b)

    cond = np.linalg.cond(A, 2)
    condition.append(cond)
    cond_inf = np.linalg.cond(A, 2)
    condition_inf.append(cond_inf)

    print("numero di condiz 2: ", cond, " \n numero di condiz inf: ", cond_inf)

    x= np.linalg.solve(A, b)

    err= np.linalg.norm(x - x_true)/np.linalg.norm(x_true)
    print("errore: ", err)
    error.append(err)


    print("x:", x)

plt.xlabel('dimension')
plt.ylabel('error')
plt.plot(n_all, condition, label='condition', color='red' )
plt.plot(n_all, condition_inf, label='condition_inf',  color='blue' )
plt.plot(n_all, error, label='error', color='green' )
plt.legend(loc='upper right')
plt.show()
```

```python
#EXERCISE 3

A=np.array([[4,2], [1,3]])
B=np.array([[4,2], [2,1]])

rank_A = np.linalg.matrix_rank(A)
eigenvalues_A, eigvecA = np.linalg.eig(A)

# Calculate rank and eigenvalues for B
rank_B = np.linalg.matrix_rank(B)
eigenvalues_B, eigvecB = np.linalg.eig(B)

print("Matrix A:")
print("Rank:", rank_A)
print("Eigenvalues:", eigenvalues_A)

print("\nMatrix B:")
print("Rank:", rank_B)
print("Eigenvalues:", eigenvalues_B)
'''
Matrix A is full rank because it has a rank of 2, and it is a 2x2 matrix.
Matrix B is not full rank because it has a rank of 1, which is less than the minimum of its 2 rows and 2 columns.
'''
```
